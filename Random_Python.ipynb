```python
import pandas as pd
import datetime as dt
import pandas_datareader.data as web
import plotly.express as px
import plotly.graph_objects as go
```

##Getting The Data

```python
start = dt.datetime(2013,1,1)
end = dt.datetime.now()

stocks = web.DataReader(['FB','AMZN', 'AAPL', 'NFLX', 'GOOGL', 'MSFT'], 'yahoo', start, end)
```

```python
stocks_close = pd.DataFrame(web.DataReader(['FB','AMZN', 'AAPL', 'NFLX', 'GOOGL', 'MSFT'], 'yahoo', start, end)['Close'])
```

##Area Chart

```python
area_chart = px.area(x=stocks_close.index, y=stocks_close.FB, title = 'FACEBOOK SHARE PRICE (2013-2020)')

area_chart.update_xaxes(title_text = 'Date')
area_chart.update_yaxes(title_text = 'FB Close Price', tickprefix = '$')
area_chart.update_layout(showlegend = False)

area_chart.show()
```

```python
c_area = px.area(x=stocks_close.index, y=stocks_close.FB, title = 'FACEBOOK SHARE PRICE (2013-2020)')

c_area.update_xaxes(
    title_text = 'Date',
    rangeslider_visible = True,
    rangeselector = dict(
        buttons = list([
            dict(count = 1, label = '1M', step = 'month', stepmode = 'backward'),
            dict(count = 6, label = '6M', step = 'month', stepmode = 'backward'),
            dict(count = 1, label = 'YTD', step = 'year', stepmode = 'todate'),
            dict(count = 1, label = '1Y', step = 'year', stepmode = 'backward'),
            dict(step = 'all')])))

c_area.update_yaxes(title_text = 'FB Close Price', tickprefix = '$')
c_area.update_layout(showlegend = False,
    title = {
        'text': 'FACEBOOK SHARE PRICE (2013-2020)',
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

c_area.show()
```

###The squad dataset contains two subsets, train and validation. If preferred, we can specify that we’d like to use just one of these with: 

split = 'train' or 'validation'

```python
dataset = datasets.load_dataset('squad', streaming=True)
```

```python
dataset
```

###To see the dataset size, description, and features, we can access the dataset_size, description, and features attributes respectively.

```python
dataset.keys()
```

```python
# both 'train' and 'validation' will output the same info
print(dataset['train'].dataset_size)
print(dataset['train'].description)
print(dataset['train'].features)
```

###When using streaming = True, we need to iterate over the dataset object for the samples to be loaded

```python
for sample in dataset['train']:
  print(sample)
  break
```

###When using streaming = False, we can use the dataset object like a normal dataset

```python
dataset['train'][0]  # only when 'streaming=False'
```

```python
for i, sample in enumerate(dataset['train']):
    print(sample['answers'])
    if i > 4: break
```

###When streaming is set to true a new dataset is created and we need to include all the features in the map function so as to not loose information, but when streaming is false we can just put the new feature columns in the map function as the rest of the features will be retained.

```python
dataset['train'] = dataset['train'].map(
    lambda x: {
        'id': x['id'],
        'context': x['context'],
        'answers': {
            **x['answers'],
            **{'answer_end': [x['answers']['answer_start'][0] + len(x['answers']['text'][0])]}
        },
        'question': x['question'],
        'title': x['title']
    }
)
```

```python
for i, sample in enumerate(dataset['train']):
    print(sample['answers'])
    if i > 4: break
```

###Testing with streaming = False

```python
dataset = datasets.load_dataset('squad', streaming=False)

dataset['train'] = dataset['train'].map(
    lambda x: {
        'answers': {
            **x['answers'],
            **{'answer_end': [x['answers']['answer_start'][0] + len(x['answers']['text'][0])]}
        }
    }
)
```

```python
dataset['train'][0]
```

###BERT Tokenization

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

dataset['train'] = dataset['train'].map(
    lambda x: tokenizer(
            x['question'], x['context'],
            max_length=512, padding='max_length',
            truncation=True
        ), batched=True, batch_size=32
)
```

```python
dataset['train']
```

###Renaming Columns

```python
dataset['train'] = dataset['train'].rename_column('title', 'topic')
dataset['train']
```

###Filtering out Rows (we cannot use filter when streaming=True)

```python
dataset['train'] = dataset['train'].filter(
    lambda x: x['topic'] != 'University_of_Notre_Dame'
)
```

###Filtering/Removing Columns

```python
dataset['train'] = dataset['train'].remove_columns([
    'answers', 'context', 'id', 'question', 'topic'
])
dataset['train']
```

##Working with Anime API(https://jikan.moe/)

```python
import urllib.request
import json
import requests
from PIL import Image
from IPython.display import Image as img
```

```python
query = input("Please Enter the Anime you want to search: ")
response = requests.get(f'https://api.jikan.moe/v3/search/anime?q={query}')
data = json.loads(response.text)
for idx, anime in enumerate(data['results']):
  print(f"{idx}: {anime['title']}")
```

```python
val = int(input("Enter the Sno. of Anime you want information on: "))
anime = data['results'][val]
if data['results'][val]['airing']:
  print(f"Name of the Anime: {anime['title']}\nLink to poster: {anime['image_url']}\nSynopsis: {anime['synopsis']}\nRating: {anime['score']}\nStart Date: {anime['start_date'][:10]}\nEnd Date: N/A (Still Airing)\nNumber of Episodes: {anime['episodes']}\nMore Details: {anime['url']}")
else:
  print(f"Name of the Anime: {anime['title']}\nLink to poster: {anime['image_url']}\nSynopsis: {anime['synopsis']}\nRating: {anime['score']}\nStart Date: {anime['start_date'][:10]}\nEnd Date: {anime['end_date'][:10]}\nNumber of Episodes: {anime['episodes']}\nMore Details: {anime['url']}")
```
